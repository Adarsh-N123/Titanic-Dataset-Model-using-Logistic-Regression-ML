```python
fin_embark = pd.get_dummies(df["Embarked"],drop_first=True)

df.drop(["Name","Ticket","Embarked","PassengerId"],axis=1,inplace=True)
fin_embark
```

```python
df.head()
```

```python
df = pd.concat([df,fin_embark],axis=1)


df = pd.get_dummies(df, columns = ["Sex"])

df = df.drop(["Sex_female"],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop("Survived",axis=1),df["Survived"],test_size=0.3,random_state=101)
x_train.head()
```

```python

sns.heatmap(df.isnull(),yticklabels=False)
```

```python
Model = LogisticRegression()
Model.fit(x_train,y_train)
```

```python
predictions = Model.predict(x_test)
predictions
```

```python
ConfusionMatrix = confusion_matrix(y_test,predictions)
AcuuracyScore = accuracy_score(y_test,predictions)
ClassificationReport = classification_report(y_test,predictions)
ConfusionMatrix 
```
